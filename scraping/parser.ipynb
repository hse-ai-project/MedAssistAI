```python
import requests
import time
import pandas as pd
from bs4 import BeautifulSoup
from matplotlib import pyplot as plt
import seaborn as sns
sns.set()
```

```python
url = 'https://klinikabudzdorov.ru/diseases/'
pages = requests.get(url)
soup = BeautifulSoup(pages.text, 'html.parser')
diseases = soup.findAll(attrs={'class': 'b-diseases-list__items'})
```

```python
dis_url = {}

for diseases_letter in diseases:
    for disease in diseases_letter.findAll('a'):
        dis_url[disease.get_text()] = disease.attrs['href']
```

```python
i = 0
N = len(dis_url.keys())
result = pd.DataFrame()
for disease, path in dis_url.items():
    url = 'https://klinikabudzdorov.ru/' + path
    try:
        page = requests.get(url)
    except:
        time.sleep(7)
        page = requests.get(url)
    soup = BeautifulSoup(page.text, 'html.parser')
    
    label = disease

    try:
        description = soup.findAll(attrs={'class': 'b-disease-intro__box'})[0].get_text()
    except:
        description = None
    
    try:
        category = soup.findAll(attrs={'class': 'b-disease-intro__info'})[0].findAll('span')[0].get_text()
        reasons = soup.findAll(attrs={'class': 'b-content__content text'})[0].findAll('p')
        reasons = [reason.get_text() for reason in reasons]
    except:
        reasons = None
    try:
        not_reasons = soup.findAll(attrs={'class': 'b-text-attention'})[0].findAll('p')[0].get_text()
        reasons = '\n'.join([reason for reason in reasons if reason not in not_reasons])
    except:
        reasons = reasons
    
    try:
        types = soup.findAll(attrs={'class': 'b-content__content text'})[1].findAll('p')
        types = '\n'.join([elem.get_text() for elem in types])
    except:
        types = None

    try: 
        symptoms = soup.findAll(attrs={'class': 'b-content__content text'})[2].findAll('p')
        symptoms = '\n'.join([elem.get_text() for elem in symptoms])
    except:
        symptoms = None

    try:
        hards = soup.findAll(attrs={'class': 'b-content__content text'})[3].findAll('p')
        hards = '\n'.join([elem.get_text() for elem in hards])
    except:
        hards = None

    try:    
        diagnostics = soup.findAll(attrs={'class': 'b-content__content text'})[4].findAll(['p', 'ul', 'ol'])
        diagnostics = '\n'.join([elem.get_text() for elem in diagnostics])
    except:
        diagnostics = None

    try:
        treatment = soup.findAll(attrs={'class': 'b-content__content text'})[5].findAll(['p', 'ul', 'ol'])
        treatment = '\n'.join([elem.get_text() for elem in treatment])
    except:
        treatment = None

    try:    
        profilactics = soup.findAll(attrs={'class': 'b-content__content text'})[6].findAll(['p', 'ul', 'ol'])
        profilactics = '\n'.join([elem.get_text() for elem in profilactics])
    except:
        profilactics = None

    try:    
        rehabilitations = soup.findAll(attrs={'class': 'b-content__content text'})[7].findAll(['p', 'ul', 'ol'])
        rehabilitations = '\n'.join([elem.get_text() for elem in rehabilitations])
    except:
        rehabilitations = None

    temp_df = pd.DataFrame([label, description, category, reasons, types, symptoms, hards, diagnostics, treatment, profilactics, rehabilitations]).T
    temp_df.columns = ['label', 'description', 'category', 'reasons', 'types', 'symptoms', 'hards', 'diagnostics', 'treatment', 'profilactics', 'rehabilitations']
    result = pd.concat([result, temp_df], axis=0)
    i += 1
    print(i / N * 100)
    time.sleep(1)
```

```python
result.to_csv('MedDataset.csv', index=False)
```

```python
result.reset_index(drop=True)
```
